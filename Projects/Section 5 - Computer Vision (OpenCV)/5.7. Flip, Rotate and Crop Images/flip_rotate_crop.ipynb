```python
import cv2
import matplotlib.pyplot as plt
```

```python
bgr_image = cv2.imread("original_image.png")
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# shape is in the format (height, width, channels)
print("BGR image shape:", bgr_image.shape)

# Giải thích về: [[[...], [...]], [[...], [...]]]
# + [...] (Lớp ngoài cùng) -> Đại diện cho cả bức ảnh (toàn bộ dữ liệu ảnh).
# + [[...], [...]] (Lớp thứ hai) -> Đại diện cho các dòng (chiều cao - height) của ảnh.
# + [[[...], [...]], [[...], [...]]] (Lớp thứ ba) -> Đại diện cho các cột (chiều rộng - width) của ảnh.
# + N1, N2, N3 (Lõi cuối cùng) -> Đại diện cho các kênh màu (channels) của ảnh, thường là BGR (Blue, Green, Red) trong OpenCV.
print(bgr_image)
```

```python
# Flip the image
flipped_image_vertical = cv2.flip(rgb_image, 0) # Vertical flip
flipped_image_horizontal = cv2.flip(rgb_image, 1) # Horizontal flip
flipped_image_both = cv2.flip(rgb_image, -1) # Both axes flip

# Display the images
titles = ['Original Image', 'Horizontal Flip', 'Vertical Flip', 'Both Axes Flip']
plt.figure(figsize=(10, 8))
images = [rgb_image, flipped_image_horizontal, flipped_image_vertical, flipped_image_both]
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
```

```python
# Rotate the image
rotated_90_clockwise = cv2.rotate(rgb_image, cv2.ROTATE_90_CLOCKWISE) # Xoay ảnh 90 độ về phía bên PHẢI (theo chiều kim đồng hồ)
rotated_180 = cv2.rotate(rgb_image, cv2.ROTATE_180) # Xoay ảnh 180 độ

# Xoay ảnh 90 độ về phía bên TRÁI (Ngược chiều kim đồng hồ)
rotated_90_counterclockwise = cv2.rotate(rgb_image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Display the images
titles = ['Original Image', '90° Clockwise', '90° Counterclockwise', '180° Rotation']
plt.figure(figsize=(10, 8))
images = [rgb_image, rotated_90_clockwise, rotated_90_counterclockwise, rotated_180]
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
```

```python
print("Original image shape:", rgb_image.shape)

# Crop the image (numpy slicing): TRỤC Y TRƯỚC - TRỤC X SAU (Height trước, Width sau)
x_start, y_start = 400, 300
x_end, y_end = 800, 700
cropped_image = rgb_image[y_start:y_end, x_start:x_end] # Crop vùng từ (y1:y2, x1:x2)
print("Cropped image shape:", cropped_image.shape)

# Display the original image and cropped image
plt.figure(figsize=(10, 5))
titles = ['Original Image', 'Cropped Image']
images = [rgb_image, cropped_image]
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
```
