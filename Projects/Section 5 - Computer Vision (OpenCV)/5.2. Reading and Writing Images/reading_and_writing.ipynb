```python
import cv2
```

```python
# Đọc ảnh từ file
image = cv2.imread('./mountain.jpg')

# Kiểm tra xem ảnh đã được đọc thành công hay chưa
if image is None:
    print("Error: Could not read the image")
else:
    print("Image read successfully!")

    # Hiển thị ảnh trong một cửa sổ
    cv2.imshow('Mountain Image', image)
    cv2.waitKey(0)  # Chờ nhấn phím bất kỳ để đóng cửa sổ
    cv2.destroyAllWindows()

    # Ghi ảnh ra file mới
    image_new = image + 30  # Tăng độ sáng của ảnh
    cv2.imwrite('mountain_copy.jpg', image_new)
    print("Image written successfully as 'mountain_copy.jpg'.")
```
