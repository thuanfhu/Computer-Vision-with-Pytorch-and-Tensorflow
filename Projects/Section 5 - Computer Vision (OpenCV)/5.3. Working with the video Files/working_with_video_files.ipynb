```python
import cv2
```

```python
#  Load a video from file: Chỉ tạo ra một kết nối (file handle) đến file video và trỏ đến frame đầu tiên (frame 0)
video = cv2.VideoCapture("./mountain.mp4")
```

```python
# Display the video frame by frame
# Lệnh video.read() đọc từng frame một (lazy loading), xử lý xong, frame đó bị ghi đè bởi frame tiếp theo ở vòng lặp sau
while True:
    ret, frame = video.read() # ret is a boolean indicating if the frame was read successfully
    if not ret:
        break
    cv2.imshow("Video", frame) # frame is numpy array representing the frame
    if cv2.waitKey(25) & 0xFF == ord('q'): # 25 ms delay between frames
        break # Press 'q' to exit the video early

# T0 ---> f1, wait 25 ms
# T1 ---> f2, wait 25 ms...
# Tính toán FPS: 1000ms / 25ms = 40 frames/giây

video.release() # Release the video object
cv2.destroyAllWindows() # Close all OpenCV windows
```

```python
# Write a video
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Define the codec (mp4v, XVID, H264, HEVC, etc.)

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))  # Get the width of the frames
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))  # Get the height of the frames

out = cv2.VideoWriter('./output.mp4', fourcc, 30, (width, height)) # Create VideoWriter object

while True:
    ret, frame = video.read()
    if not ret:
        break
    out.write(frame) # Write the frame to the output file

video.release() # Release the input video object
out.release() # Release the output video object
cv2.destroyAllWindows() # Close all OpenCV windows
```
