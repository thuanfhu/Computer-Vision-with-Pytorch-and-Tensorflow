```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image in grayscale, cv2.IMREAD_GRAYSCALE ensures the image is read as a single channel
image = cv2.imread('filter_image.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
```

```python
# Define a sharpening kernel
x = -10
horizontal_kernel = np.array([[-1, -1, -1], [x, x, x], [1, 1, 1]])
vertical_kernel = np.array([[-1, x, 1], [-1, x, 1], [-1, x, 1]])
diagonal_kernel1 = np.array([[x, -1, -1], [1, x, -1], [1, 1, x]])
diagonal_kernel2 = np.array([[-1, -1, x], [-1, x, 1], [x, 1, 1]])

# Apply the filter using cv2.filter2D
# ddepth=-1 means the output image will have the same depth as the source image
horizontal_filtered_image = cv2.filter2D(src=image, ddepth=-1, kernel=horizontal_kernel)
vertical_filtered_image = cv2.filter2D(src=image, ddepth=-1, kernel=vertical_kernel)
diagonal_filtered_image1 = cv2.filter2D(src=image, ddepth=-1, kernel=diagonal_kernel1)
diagonal_filtered_image2 = cv2.filter2D(src=image, ddepth=-1, kernel=diagonal_kernel2)

# Display the original and horizontally filtered images side by side
plt.figure(figsize=(15, 5))
titles = ['Original Image', 'Horizontally Image', 'Vertically Image', 'Diagonally Image 1', 'Diagonally Image 2']
images = [image, horizontal_filtered_image, vertical_filtered_image, diagonal_filtered_image1, diagonal_filtered_image2]
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
```
