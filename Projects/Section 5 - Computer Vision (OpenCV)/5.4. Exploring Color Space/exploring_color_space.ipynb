```python
import cv2
```

```python
# OpenCV mặc định sử dụng không gian màu BGR (do đây là chuẩn kĩ thuật phổ biến những năm 2000)
bgr_image = cv2.imread("original_image.jpg")
```

```python
# Convert BGR to Grayscale
gray_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY)
cv2.imwrite("gray_image.jpg", gray_image)
# -> Grayscale thường được sử dụng trong các ứng dụng như nhận diện khuôn mặt, phát hiện cạnh, và xử lý ảnh y tế, nơi mà thông tin về độ sáng quan trọng hơn thông tin về màu sắc.
```

```python
# Convert BGR to HSV
hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)
cv2.imwrite("hsv_image.jpg", hsv_image)
# -> HSV chủ yếu sử dụng trong các ứng dụng liên quan đến xử lý màu sắc như phát hiện màu sắc, phân đoạn ảnh dựa trên màu sắc, điều chỉnh màu sắc trong ảnh và video, v.v.
```

```python
# Convert BGR to LAB
lab_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2LAB)
cv2.imwrite("lab_image.jpg", lab_image)
# -> LAB thường được sử dụng trong các ứng dụng yêu cầu sự chính xác về màu sắc cao như in ấn, thiết kế đồ họa, và phân tích hình ảnh chuyên sâu.
```

```python
# Convert BGR to RGB
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
cv2.imwrite("bgr_image.jpg", rgb_image) # Chỗ này output đang là BGR image
# -> RGB là không gian màu phổ biến nhất được sử dụng trong các thiết bị hiển thị
```
