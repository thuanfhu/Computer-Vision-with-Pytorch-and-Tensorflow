```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Function: Show original images
def show_images(images, titles, cmap=None):
    plt.figure(figsize=(15, 10))
    for i in range(len(images)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i], cmap=cmap)
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

```python
# Load images
gaussian_noise_image = cv2.imread('gaussian_noise_image.png')
salt_and_pepper_noise_image = cv2.imread('salt_pepper_noise_image.png')
original_image = cv2.imread('original_image.png')

# Convert BGR to RGB
gaussian_noise_image = cv2.cvtColor(gaussian_noise_image, cv2.COLOR_BGR2RGB)
salt_and_pepper_noise_image = cv2.cvtColor(salt_and_pepper_noise_image, cv2.COLOR_BGR2RGB)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

# Display the images
images = [original_image, gaussian_noise_image, salt_and_pepper_noise_image]
titles = ['Original Image', 'Gaussian Noise Image', 'Salt and Pepper Noise Image']
show_images(images,titles)
```

```python
# Blur Filters Implementation
kernel_size = 11

# 1. Average Blur: Lm m u
average = cv2.blur(original_image, (kernel_size, kernel_size))

# 2. Gaussian Blur: Lm m mm mi
# kernel_size phi l s l
# 0 ngha l  OpenCV t tnh  lch chun (sigma)
gaussian = cv2.GaussianBlur(gaussian_noise_image, (kernel_size, kernel_size), 0)

# 3. Median Blur: Chuyn tr nhiu mui tiu
# kernel_size phi l s l
median = cv2.medianBlur(salt_and_pepper_noise_image, kernel_size)

# Display the blurred images
images = [average, gaussian, median]
titles = ['Average Blur', 'Gaussian Blur', 'Median Blur']
show_images(images,titles)
```
